# Making a scatterplot in plotnine

Scatter plots are simple in plotnine! After setting your x and y axes, you use `geom_point()` to draw a point for each of your datapoints.

```python
import pandas as pd
from plotnine import *

df = pd.read_csv('countries.csv')
df.head(2)
```

```python
(
    ggplot(df)
    + aes(x='gdp_per_capita', y='life_expectancy')
    + geom_point()
)
```
