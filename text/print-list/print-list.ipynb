# Print a list of fonts supported by plotnine

## Setting fonts for plotnine

When making a graph with plotnine, you can use `element_text(family=...)` to pick a font.

```python
import pandas as pd
from plotnine import *

df = pd.read_csv('countries.csv')

(
    ggplot(df)
    + aes(x='gdp_per_capita', y='life_expectancy')
    + geom_point()
    + labs(title="Life Expectancy vs GDP (per capita)")
    + theme(plot_title=element_text(family='Frutiger Bold'))
)
```

# List of fonts supported by plotnine

Sometimes you don't know the "real" name of a font, and when you use `family=` you might get an error like `findfont: Font family ['Frutiger'] not found. Falling back to DejaVu Sans.`.

What fonts are available? Let's ask maptlotlib!

```python
import matplotlib.font_manager
from IPython.core.display import HTML

def make_html(fontname):
    return "<p>{font}: <span style='font-family:{font}; font-size: 24px;'>{font}</p>".format(font=fontname)

font_names = sorted(set([f.name for f in matplotlib.font_manager.fontManager.ttflist]))
code = "\n".join([make_html(font) for font in font_names])

HTML("<div style='column-count: 2;'>{}</div>".format(code))
```
